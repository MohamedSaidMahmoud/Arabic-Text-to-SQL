```python
import gradio as gr
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel

# Load the model and tokenizer
base_model = AutoModelForCausalLM.from_pretrained("motherduckdb/DuckDB-NSQL-7B-v0.1")
model = PeftModel.from_pretrained(base_model, "Rahmaa33/MotherDuckTEXT2SQLLArabicc")
tokenizer = AutoTokenizer.from_pretrained("Rahmaa33/MotherDuckTEXT2SQLLArabicc")

# Define the function to process the input and generate SQL query
def generate_sql(query: str):
    prompt = f"""### Instructions:
    Given the following question, generate an SQL query for the database department_management.

    Question: {query}

    SQL:"""

    # Tokenize the input and generate the SQL query
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(**inputs, max_new_tokens=300)
    
    # Decode the output and return the SQL query
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Create a Gradio interface for the model
interface = gr.Interface(
    fn=generate_sql, 
    inputs=gr.Textbox(label="Enter Your Question"), 
    outputs=gr.Textbox(label="Generated SQL Query"), 
    live=True
)

# Launch the interface
interface.launch()
```
